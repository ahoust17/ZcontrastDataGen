```python

import numpy as np
import matplotlib.pyplot as plt
%matplotlib ipympl

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import Input, Conv2D, LeakyReLU, BatchNormalization, UpSampling2D, Concatenate, MaxPooling2D
from tensorflow.keras import Model
from keras.callbacks import TensorBoard
from sklearn.utils import class_weight
```

```python
n = 3

model_names = ['WS2','HBN','cubic']
dataset_names = ['WS2','HBN','cubic']
fig, axs = plt.subplots(9,3, sharex = True, sharey = True)

i = 0
for model_name in model_names:
    model = keras.models.load_model('./model_saves/' + model_name + '_trained')
    

    for dataset_name in dataset_names:
        dataset = np.load('/Users/austin/tensorflow_env/COSC_525_DNN/Final/datasets/' + dataset_name + '_train_test.npz')
        # data management
        images = dataset['X_train'][:,:,0,:,:]
        labels = dataset['y_train']
        images_test = dataset['X_test'][:,:,0,:,:]
        labels_test = dataset['y_test']

        images = images.reshape((1280, 256, 256, 1))
        labels = labels.reshape((1280, 256, 256, 1))
        images_test = images_test.reshape((320, 256, 256, 1))
        labels_test = labels_test.reshape((320, 256, 256, 1))

        labels_onehot = to_categorical(labels)
        labels_test_onehot = to_categorical(labels_test)

        class_freq = np.unique(labels, return_counts=True)[1]
        class_weights = class_weight.compute_class_weight(class_weight = 'balanced', classes = np.unique(labels), y = labels.flatten())
        class_weights = dict(enumerate(class_weights))

        sample_weights = np.zeros_like(labels)
        for ig in range(len(class_weights)):
            sample_weights[labels == ig] = class_weights[ig]

        # make predictions
        predictions = model.predict(images_test)
        categorical_labels = np.argmax(np.round(predictions), axis = -1)

        # plot results
        
        axs[i,0].set_title(dataset_name + ' Dataset')
        axs[i,0].imshow(images_test[n], cmap = 'gray')

        axs[i,1].set_title(model_name + ' Model Pred.')
        axs[i,1].imshow(categorical_labels[n], cmap = 'jet')

        axs[i,2].set_title('Ground Truth')
        axs[i,2].imshow(labels_test[n], cmap = 'jet')

        axs[i,0].set_axis_off()
        axs[i,1].set_axis_off()
        axs[i,2].set_axis_off()
        i = i + 1

fig.show()
```
