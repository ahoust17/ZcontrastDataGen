```python
dataset = np.load('/Users/austin/Desktop/Projects/PdSeM_train_test.npz')
images = dataset['X_train'].reshape(-1,1,256,256)
labels = dataset['y_train'].reshape(-1,256,256)
images_test = dataset['X_test'].reshape(-1,1,256,256)
labels_test = dataset['y_test'].reshape(-1,256,256)
```

```python
print(images.shape, labels.shape)
print(images_test.shape, labels_test.shape)
```

```python
n = 5 # number of images to plot

n = n + 1
fig = plt.figure( figsize=(30, 8))
for i in range(1, n):   
    ax = fig.add_subplot(2, n, i)
    ax.imshow(images[i-1,0,:,:], cmap='gray')
    ax.set_title('Augmented image ' + str(i))
    ax.grid(alpha = 0.5)
    ax = fig.add_subplot(2, n, i+n)
    ax.imshow(labels[i-1], interpolation='Gaussian', cmap='jet')
    ax.set_title('Ground truth ' + str(i))
    ax.grid(alpha=0.75)
```
